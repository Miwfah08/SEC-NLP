```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns  # tambahkan import seaborn untuk menggunakan sns.barplot

# Pisahkan data berdasarkan keyword
df_kampus_mengajar = df[df['normalized_x'].str.contains('kampus mengajar|kamjar', case=False, na=False)]

# Filter data untuk hanya mengambil sentimen negatif
df_negatif = df_kampus_mengajar[df_kampus_mengajar['manual'] == 'negatif']

# Gabungkan semua teks dari sentimen negatif
text_negatif = ' '.join(df_negatif['normalized_x'].dropna().tolist())

# List kata-kata yang perlu di-exclude
exclude_words = ['kamjar', 'banget', 'sama', 'iya', 'kamu', 'jadi', 'apa', 'ikut', 'mau', 'pas', 'kayak', 'tahu', 'orang', 'padahal', 'buat', 'hari', 'mana', 'masuk']

# Fungsi untuk menghapus kata-kata yang perlu di-exclude
def remove_words(text, words):
    for word in words:
        text = text.replace(word, '')
    return text

# Menghapus kata-kata yang perlu di-exclude dari teks
text_negatif_cleaned = remove_words(text_negatif, exclude_words)

# Buat WordCloud dan ekstrak frekuensi kata
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(text_negatif_cleaned)
word_freq = wordcloud.words_

# Konversi ke DataFrame
word_freq_df = pd.DataFrame(list(word_freq.items()), columns=['Word', 'Frequency'])

# Plot horizontal bar chart
plt.figure(figsize=(15, 7))
word_freq_df = word_freq_df.head(15)  # Menampilkan 20 kata yang paling sering muncul
sns.barplot(x='Frequency', y='Word', data=word_freq_df, palette='Reds_d')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.title('Top 15 Words in Negative Sentiments for "Kampus Mengajar" and "Kamjar"')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pisahkan data berdasarkan keyword
df_kampus_mengajar = df[df['normalized_x'].str.contains('kampus mengajar|kamjar', case=False, na=False)]

# Filter data untuk hanya mengambil sentimen negatif
df_negatif = df_kampus_mengajar[df_kampus_mengajar['manual'] == 'negatif']

# Daftar kata spesifik yang akan dihitung frekuensinya
keywords = ['bbh', 'konversi', 'sedih', 'nangis', 'capek', 'lelah', 'jauh']

# Hitung frekuensi kemunculan kata-kata spesifik
keyword_counts = {keyword: df_negatif['normalized_x'].str.contains(keyword, case=False, na=False).sum() for keyword in keywords}

# Konversi ke DataFrame untuk visualisasi
keyword_counts_df = pd.DataFrame(list(keyword_counts.items()), columns=['Keyword', 'Frequency'])

# Plot barchart
plt.figure(figsize=(10, 6))
sns.barplot(x='Keyword', y='Frequency', data=keyword_counts_df, palette='Blues_d')
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.title('Frequency of Specific Words in Negative Sentiments for "Kampus Mengajar" and "Kamjar"')
plt.show()
```

```python
# Pisahkan data berdasarkan keyword
df_iisma = df[df['normalized_x'].str.contains('iisma', case=False, na=False)]
df_kampus_mengajar = df[df['normalized_x'].str.contains('kampus mengajar|kamjar', case=False, na=False)]
df_studpen = df[df['normalized_x'].str.contains('studi independen|studpen', case=False, na=False)]
df_msib = df[df['normalized_x'].str.contains('magang msib|msib', case=False, na=False)]
df_konversi = df[df['normalized_x'].str.contains('mbkm|kampus merdeka', case=False, na=False)]


# Hitung proporsi label untuk setiap keyword
prop_iisma = df_iisma['manual'].value_counts(normalize=True).sort_index()
prop_kampus_mengajar = df_kampus_mengajar['manual'].value_counts(normalize=True).sort_index()
prop_studpen = df_studpen['manual'].value_counts(normalize=True).sort_index()
prop_msib = df_msib['manual'].value_counts(normalize=True).sort_index()
prop_konversi = df_konversi['manual'].value_counts(normalize=True).sort_index()


# Gabungkan proporsi ke dalam satu dataframe
prop_df = pd.DataFrame({
    'iisma': prop_iisma,
    'kampus_mengajar': prop_kampus_mengajar,
    'studpen' : prop_studpen,
    "msib" : prop_msib,
    "kampus merdeka": prop_konversi
}).fillna(0).reset_index()

# Plot barplot
prop_df_melt = prop_df.melt(id_vars='index', var_name='keyword', value_name='proportion')

# Warna pastel untuk setiap label
pastel_colors = ['#537ABB','#7F62A2','#4AACC5']

# Membuat plot lebih besar
plt.figure(figsize=(12, 8))  # Atur ukuran figure (width, height)
sns.barplot(x='keyword', y='proportion', hue='index', data=prop_df_melt, palette=pastel_colors)
plt.xlabel('Keyword')
plt.ylabel('Proportion')
plt.title('Proportion of Labels for Keywords')
plt.legend(title='Label')
plt.show()
```
